### Conversion de date_derniere_maintenance en type datetime

```python
df['date_derniere_maintenance'] = pd.to_datetime(df['date_derniere_maintenance'])
df['date_derniere_maintenance']
```

```python
pd.to_datetime('today')
```

```python
# Calculer le nombre de jours depuis la dernière maintenance
df['jours_depuis_dern_maintenance'] = (pd.to_datetime('today') - df['date_derniere_maintenance']).dt.days
```

```python
df.head()
```

```python
# Convertir 'etat_actuel' en binaire
df['etat_actuel_binaire'] = df['etat_actuel'].apply(lambda x: 1 if x == 'nécessite une maintenance' else 0)
```

```python
# Sélectionner les caractéristiques numériques
numeric_features = df[['age', 'heures_fonctionnement', 'temperature', 'vibration', 'pression', 'jours_depuis_dern_maintenance']]
correlation_matrix = numeric_features.corrwith(df['etat_actuel_binaire']).sort_values(ascending=False)

# Afficher la corrélation
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.barplot(x=correlation_matrix.values, y=correlation_matrix.index)
plt.title('Corrélation entre les caractéristiques et la variable cible')
plt.show()
```

```python
df.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
# Encoder le type de machine
df = pd.get_dummies(df, columns=['type_machine'])

# Sélectionner les caractéristiques
X = df[['age', 'heures_fonctionnement', 'temperature', 'vibration', 'pression', 'jours_depuis_dern_maintenance'] + list(df.columns[df.columns.str.startswith('type_machine_')])]
y = df['etat_actuel_binaire']

# Entraîner le modèle
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Importance des caractéristiques
importances = model.feature_importances_
feature_importance = pd.Series(importances, index=X.columns).sort_values(ascending=False)

# Afficher l'importance
print(feature_importance)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=feature_importance.values, y=feature_importance.index)
plt.title('Importance des caractéristiques')
plt.show()
```

#### D'après les résultats de l'importance des caractéristiques, nous pouvons observer les points suivants :

##### jours_depuis_dern_maintenance : Cette caractéristique a la plus grande importance (0.535923). Cela indique que le nombre de jours depuis la dernière maintenance est le facteur le plus influent pour prédire si un équipement nécessite une maintenance.
##### vibration, pression, heures_fonctionnement, age, temperature : Ces caractéristiques ont une importance modérée, suggérant qu'elles jouent un rôle significatif mais moins critique que la durée depuis la dernière maintenance.
##### type_machine : Les différentes catégories de machines ont des importances variables, avec type_machine_Niveleuse et type_machine_Grue ayant une importance légèrement plus élevée parmi les types de machines. Cependant, leur importance est relativement faible comparée aux autres caractéristiques.

```python
def plot_confusion_matrix(conf_matrix, model_name):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Fonctionnelle', 'Nécessite une maintenance'], yticklabels=['Fonctionnelle', 'Nécessite une maintenance'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix for {model_name}')
    
    plt.show()

# Calculer les matrices de confusion
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
```

```python
plot_confusion_matrix(conf_matrix_rf, "Random Forest")
plot_confusion_matrix(conf_matrix_svm, "SVM")
```
